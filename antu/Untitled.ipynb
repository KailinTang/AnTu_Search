```python
def find_no(data,num):
    i,flag = 0,False
    while i < len(data):
        if data['No'][i] == num:
            flag = True
            break
    
        i += 1
        
    if flag:
        return i
    else:
        return -1
    
```

```python
def compare(s1,s2):
    ret=[]
    for i in s1.split(' '):
        namei=i.lower()
        for j in s2.split(' '):
            scorej=0.0
            namej=j.lower()
            for k in namei:
                if k in namej:
                    scorej+=1.0
                
            if len(i) != 0:
                sc=scorej/(len(i)*1.0)
            else:
                sc=0
            
            if sc >0.5:
                ret.append(sc)
    if len(ret)>0:
        score = sum(ret)/len(ret)
    else:
        score = 0

    return score
```
